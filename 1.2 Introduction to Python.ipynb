*Python Variable*

Type of the object is defined at the runtime.


So, a Python variable name is just a name that does not have a type.
It only points to some object, and the type of the object is stored in the object itself, which cannot be changed.
Hence, Python is a strongly typed language,which means the type of an object cannot be changed because it is part of its definition.

Python determines the type of the object at runtime while performing an operation with the object.
And that is how it is a dynamically typed language.

When the reference count of an object becomes zero,the object is no longer needed,and it is ready to be garbage collected by the Python garbage collection.

```python
e = 0
o = 0

for i in range(10):
    if i % 2 == 0:
        e += 1
        print("Even Number")
    else:
        o +=1
        print("Odd Number")

print("No. of Odd Numbers = " + str(o))
print("No. of Even Numbers = " + str(e))
```
